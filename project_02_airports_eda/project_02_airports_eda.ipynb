```python
# --- Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- Step 2: Load dataset
url = "https://raw.githubusercontent.com/jpatokal/openflights/master/data/airports.dat"
df = pd.read_csv(url, header=None)

# --- Step 3: Rename columns
df.columns = [
    "Airport_ID", "Name", "City", "Country",
    "IATA", "ICAO", "Latitude", "Longitude",
    "Altitude", "Timezone", "DST", "Tz_database_time_zone", "Type", "Source"
]

# --- Step 4: Quick check
print("Dataset shape:", df.shape)
df.head()
```

##Visualizations

###1. Top 10 Cities with Most Airports

```python
plt.figure(figsize=(10,5))
sns.countplot(
    y=df["City"],
    order=df["City"].value_counts().head(10).index,
    palette="viridis"
)
plt.title("Top 10 Cities with Most Airports")
plt.xlabel("Number of Airports")
plt.ylabel("City")
plt.show()
```

Insight: Major cities host the highest number of airports, reflecting their importance in global air traffic.

###2. Distribution of Airport Altitudes

```python
plt.figure(figsize=(10,5))
plt.hist(df["Altitude"], bins=50, edgecolor="black")
plt.title("Distribution of Airport Altitudes")
plt.xlabel("Altitude (feet)")
plt.ylabel("Frequency")
plt.show()
```

Insight: Most airports are located at low altitudes, with only a few rare cases at very high elevations.


###3. World Distribution of Airports

```python
plt.figure(figsize=(12,6))
plt.scatter(df["Longitude"], df["Latitude"], alpha=0.3, s=10)
plt.title("World Distribution of Airports")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

Insight: Airports are globally distributed, with clear density in North America and Europe.

###4. Distribution of Airport Types

```python
plt.figure(figsize=(6,6))
df["Type"].value_counts().plot.pie(
    autopct="%1.1f%%",
    startangle=90,
    colors=sns.color_palette("pastel")
)
plt.title("Distribution of Airport Types")
plt.ylabel("")
plt.show()
```

Insight: The majority of entries are classified as major airports, with a smaller proportion belonging to private or smaller airfields.


Conclusion:
-----------
- Airports are heavily concentrated in key global cities.
- Most airports exist at low altitudes.
- Geographical distribution shows strong clusters in North America and Europe.
- The dataset is dominated by major airports compared to other types.

This project demonstrates how Exploratory Data Analysis (EDA) can reveal useful insights in aviation datasets.
